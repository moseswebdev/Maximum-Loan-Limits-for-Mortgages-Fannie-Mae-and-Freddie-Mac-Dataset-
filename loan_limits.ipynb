```python
# Remove currency symbols on one unit limit and convert to numeric
df['One Unit Limit'] = df['One Unit Limit'].replace('[\\$,]', '', regex=True).astype(float)

# Group by County Name and State, then calculate mean
result = df.groupby(['County Name', 'State'])['One Unit Limit'].mean()
print(result)
```

```python
# Remove currency symbols in Two Unit Limit and convert to numeric
df['Two Unit Limit'] = df['Two Unit Limit'].replace('[\\$,]', '', regex=True).astype(float)

# Group by County Name and State, then calculate mean
result = df.groupby(['County Name', 'State'])['Two Unit Limit'].mean()
print(result)
```

```python
# Remove currency symbols four unit limit columns and convert to numeric
df['Four-Unit Limit'] = df['Four-Unit Limit'].replace('[\\$,]', '', regex=True).astype(float)

# Group by County Name and State, then calculate mean
result = df.groupby(['County Name', 'State'])['Four-Unit Limit'].mean()
print(result)
```
